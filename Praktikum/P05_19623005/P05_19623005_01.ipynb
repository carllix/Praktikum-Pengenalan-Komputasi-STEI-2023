```python
# Mengimport library yang dibutuhkan
import pandas as pd

# Membaca data P05-03.csv
df = pd.read_csv('P05-03.csv')
df
```

```python
# 1. Menghitung dan menampilkan banyaknya rumah pada masing-masing neighbourhood group
print(df['neighbourhood_group'].value_counts())
```

```python
# 2. Menampilkan rumah dengan jumlah review di aplikasi terbanyak selama 12 bulan
print(f"Rumah dengan jumlah review di aplikasi terbanyak selama 12 bulan adalah {df.loc[df['number_of_reviews'].idxmax()]['name']}")
```

```python
# 3. Menampilkan 5 data rumah dengan jumlah malam penyewaan minimum terendah
df.sort_values(['minimum_nights'], ascending=True)[:5]
```

```python
# 4. Menampilkan Rumah yang dimiliki oleh Jayce
df.loc[df['host_name']=='Jayce']
```

```python
# 5. Menampilkan koefisien korelasi antara banyaknya review tiap bulan dengan jumlah malam penyewaan minimum. 
df['reviews_per_month'].corr(df['minimum_nights'])
```

Berdasarkan koefisien korelasi tersebut, dapat disimpulkan bahwa banyaknya review tiap bulan dengan jumlah malam penyewaan minimum tidak berkorelasi karena nilai koefisien tersebut mendekati 0.
