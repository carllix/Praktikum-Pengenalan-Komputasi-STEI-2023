```python
# Mengimport library yang dibutuhkan
import pandas as pd

# Membaca data P05-03.csv
df = pd.read_csv('P05-03.csv')
df
```

```python
# 1. Menghitung banyak data dengan jumlah malam penyewaan minimumnya 1 namun banyaknya review di aplikasi selama 12 bulan tidak 1
print(f"Banyak data dengan jumlah malam penyewaan minimumnya 1 namun banyaknya review di aplikasi selama 12 bulan tidak 1 adalah {len(df.loc[(df['minimum_nights']==1) & (df['number_of_reviews']!=1)])}")
```

```python
# 2. Menampilkan pemilik rumah dengan banyaknya hari ketersediaan rumah dalam satu tahunnya terbanyak
print(f"Pemilik rumah dengan banyaknya hari ketersediaan rumah dalam satu tahunnya terbanyak adalah {df.loc[df['availability_365'].idxmax()]['host_name']}")
```

```python
# 3. Menghitung banyak rumah di daerah (neighbourhood group) Central Region atau East Region dengan harga rumah per malam di bawah 100

print(f"Banyak rumah di daerah (neighbourhood group) Central Region atau East Region dengan harga rumah per malam di bawah 100 adalah {len(df.loc[((df['neighbourhood_group']=='Central Region')|(df['neighbourhood_group']=='East Region')) & (df['price']<100)])}")
```

```python
# 4. Menampilkan banyaknya rumah dengan review per bulan di atas 1 untuk tiap jenis penyediaan ruang
reviews_per_month_diatas1 = df.loc[df['reviews_per_month']>1]
reviews_per_month_diatas1['room_type'].value_counts()
```

```python
# 5. Menampilkan 5 nama rumah berdasarkan banyaknya rumah yang disewakan oleh host terbanyak
df.sort_values(['calculated_host_listings_count'], ascending=False)[:5]
```
