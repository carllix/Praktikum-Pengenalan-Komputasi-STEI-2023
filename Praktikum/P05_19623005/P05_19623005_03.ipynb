```python
# NIM/Nama  : 19623005/Carlo Angkisan
# Tanggal   : 16 November 2023
# Deskripsi : Problem 3
              # 1. Menampilkan histogram distribusi banyaknya review di aplikasi yang memiliki jenis penyediaan ruang Private room dengan bins 30
              # 2. Menampilkan pie chart perbandingan banyaknya masing-masing jenis penyediaan ruang dengan banyaknya review diaplikasi di atas 100.
              # 3. Menampilkan scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang tidak dimiliki oleh Jay
              # 4. Menampilkan Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya hari ketersediaan rumah dalam satu tahun sebagai ukuran dari rumah yang berada di neighbourhood Novena.
              # 5. Menampilkan Stacked bar chart dari rumah yang disewakan oleh ’JoJo’, ’Elizabeth’, dan ’Jimmy’, dengan room type sebagai x dan nama pemilik rumah sebagai stacked y
```

```python
# Mengimport library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data P05-03.csv
df = pd.read_csv('P05-03.csv')
df
```

```python
# 1. Menampilkan histogram distribusi banyaknya review di aplikasi yang memiliki jenis penyediaan ruang Private room dengan bins 30
room_types_private = df.loc[df['room_type']=='Private room']
room_types_private['number_of_reviews'].plot(kind='hist', bins=30, title='Distribusi Banyaknya Review di Aplikasi Yang Memiliki Jenis Penyediaan Ruang Private Room', xlabel='Number of Reviews')

plt.show()
```

```python
# 2. Menampilkan pie chart perbandingan banyaknya masing-masing jenis penyediaan ruang dengan banyaknya review diaplikasi di atas 100.
number_of_reviews_diatas100 = df.loc[df['number_of_reviews']>100]
number_of_reviews_diatas100['room_type'].value_counts().plot(kind='pie', ylabel='', title='Perbandingan Banyaknya Masing-Masing Jenis Penyediaan Ruang Dengan Banyaknya Review Diaplikasi Di Atas 100', autopct='%1.1F')

plt.show()
```

```python
# 3. Menampilkan scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang tidak dimiliki oleh Jay
host_name_not_jay = df.loc[df['host_name']!='Jay']
host_name_not_jay.plot(kind='scatter', x='longitude', y='latitude', title='Scatter Plot Dengan Long Sebagai X dan Lat Sebagai Y dari Rumah yang Tidak Dimiliki oleh Jay')

plt.show()
```

```python
# 4. Menampilkan Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya hari ketersediaan rumah dalam satu tahun sebagai ukuran dari rumah yang berada di neighbourhood Novena.
neighbourhood_Novena = df.loc[df['neighbourhood']=='Novena']
neighbourhood_Novena.plot(kind='scatter', x='longitude', y='latitude', sizes=neighbourhood_Novena['availability_365'])

plt.show()
```

```python
# 5. Menampilkan Stacked bar chart dari rumah yang disewakan oleh ’JoJo’, ’Elizabeth’, dan ’Jimmy’, dengan room type sebagai x dan nama pemilik rumah sebagai stacked y
host_name_Jojo_Elizabeth_Jimy = df.loc[(df['host_name']=='Jojo')|(df['host_name']=='Elizabeth')|(df['host_name']=='Jimmy')]
host_name_Jojo_Elizabeth_Jimy.groupby(['room_type','host_name']).size().unstack().plot(kind='bar',stacked=True)

plt.show()
```
