```python
# Mengimport library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data
df = pd.read_csv('H05.csv')
```

```python
# 1. Menampilkan histogram distribusi nilai matematika
df['nilai_mat'].plot(kind='hist', title='Histogram Distribusi Nilai Matematika', xlabel='Nilai Matematika', ylabel='Banyak Mahasiswa', rwidth=0.7)
plt.show()
```

```python
# 2. Menampilkan diagram batang horizontal banyaknya mahasiswa masing-masing fakultas
df['fakultas'].value_counts().plot(kind='barh', title='Diagram Batang Horizontal Banyaknya Mahasiswa Masing-Masing Fakultas', xlabel='Banyak Mahasiswa', ylabel='Fakultas')
plt.show()
```

```python
# 3. Menampilkan diagram pie banyaknya mahasiswa masing-masing fakultas
df['fakultas'].value_counts().plot(kind='pie', title='Diagram Pie Banyaknya Mahasiswa Masing-Masing Fakultas', ylabel='')
plt.show()
```

```python
# 4. Menentukan fakultas dengan mahasiswa terbanyak dan diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak
```

Berdasarkan diagram batang horizontal, dapat kita lihat bahwa potongan batang terpanjang dimiliki oleh FTI. Begitu pula pada diagram pie, dapat kita lihat bahwa potongan pie terbesar dimiliki oleh FTI dengan potongan berwarna biru. Dengan demikian, dapat disimpulkan bahwa fakultas dengan mahasiswa terbanyak adalah FTI.

Menurut saya, diagram yang lebih baik dalam menampilkan fdakultas dengan mahasiswa terbanyak adalah diagram batang horizontal. Hal ini karena pada diagram batang horizontal kita dapat melihat dengan jelas perbedaan panjang batang dari masing-masing fakultas. Sedangkan, pada diagram pie dapat kita lihat bahwa potongan berwarna biru yang dimiliki oleh FTI dengan potongan berwarna oranye yang dimiliki oleh STEI memiliki besar yang hampir sama dan tidak ditampilkan besar sudut atau persentasenya, sehingga cukup sulit untuk menentukan fakultas mana yang memiliki mahasiswa terbanyak.

```python
# 5. Menampilkan scatter plot dengan nilai kimia sebagai x dan fisika sebagai y
df.plot(kind='scatter', x='nilai_kim', y='nilai_fis', xlabel='Nilai Kimia', ylabel='Nilai Fisika', title='Scatter Plot antara Nilai Kimia dan Fisika')
plt.show()
```
